# Dive Prediction - Hidden Markov Model

*Predicting Seabird Diving Behaviour from GPS data*

This notebook trains a HMM to predict seabirds' dives.

HMM' characteristics:

* *Number of modes* : 3

<div class="alert alert-info">
 ⚠️ Notebook using R
</div>

```r
library(momentuHMM)
library(pracma)
library(lubridate)
```

```r
options(stringsAsFactors = FALSE)

data_test = read.table("./data/LB_test.csv", sep = ',', header = TRUE)
data_test$datetime <- as.POSIXct(data_test$datetime, tz = 'GMT')
```

```r
## 
resolution = 5
```

```r
## Resample at Resolution
data_test_new <- data.frame()

for (id in unique(data_test$trip)){
    t <- data_test[data_test$trip == id,]
    
    idx = seq(1, nrow(t), by = resolution)
    traj <- t[idx,c("trip", "datetime", "lon", "lat")]
    
    # dives as maximum in window 
    traj$dive = sapply(idx, function(i){
            max(t$dive[i:(i+resolution-1)], na.rm = TRUE)
        })
    
    traj$gaps = sapply(idx, function(i){
        max((t$gaps[i:(i+resolution-1)] == 'True'), na.rm = TRUE)
    })
    
    
    data_test_new <- rbind(data_test_new, traj)
}
data_test_new
```

# Fit HMM

```r
# prep data for HMM
rawData <- data_test_new[,c("trip","lon","lat", "datetime", "dive", "gaps")]
colnames(rawData) <- c("ID", "lon", "lat", "datetime", "dive", "gaps")
birdData <- prepData(data=rawData, type = "LL", coordNames = c("lon", "lat"))
birdData
```

## 3 states no gaps

```r
# parameters initialization
## cluster K-Means for initialization
### STEP
clusterBird_step <- kmeans(na.omit(data.frame(birdData$step)), 3)
muS_1 <- max(clusterBird_step$centers)
muS_2 <- median(clusterBird_step$centers) 
muS_3 <- min(clusterBird_step$centers) 
sdS_1 <- sd(na.omit(birdData$step)[clusterBird_step[[1]] == which(clusterBird_step$centers == max(clusterBird_step$centers))])
sdS_2 <- sd(na.omit(birdData$step)[clusterBird_step[[1]] == which(clusterBird_step$centers == median(clusterBird_step$centers))])
sdS_3 <- sd(na.omit(birdData$step)[clusterBird_step[[1]] == which(clusterBird_step$centers == min(clusterBird_step$centers))])

### ANGLE
## for von mises
kappaA_1 <- 4
kappaA_2 <- 2
kappaA_3 <- 0.1

### ZERO MASS
zeroMass <- length(which(birdData$step == 0))/nrow(birdData) #we need to include zeroMass parameters

### FIT MODEL
stateNames <- c("fly","observe", "dive")
dist = list(step = "gamma", angle = "vm")
anglePar0 <- c(kappaA_1, kappaA_2, kappaA_3) 
stepPar0 <- c(muS_1, muS_2, muS_3, sdS_1, sdS_2, sdS_3, zeroMass, zeroMass, zeroMass)
# stepPar0 <- c(muS_1, muS_2, muS_3, sdS_1, sdS_2, sdS_3) 
```

```r
m <- fitHMM(data = birdData, nbStates = 3, dist = dist,
            Par0 = list(step = stepPar0, angle = anglePar0), stateNames = stateNames)
```

```r
data <- m$data
prob = stateProbs(m)
data$prediction = prob[, 'dive']
```

```r
write.table(data, file = paste0('./results/SV_hmm_', resolution, 's.csv'), sep = ',', row.names = FALSE)
```

```r
formula = ~ gaps
DM <- list(step = list(mean = ~ gaps, sd = ~ gaps, zeromass = ~gaps), angle = list(concentration = ~ gaps))
Par0 <- getPar0(model=m, formula=formula, DM = DM)

m <- fitHMM(data = birdData, nbStates = 3, dist = dist, DM = DM,
            Par0 =Par0$Par, beta0 = Par0$beta, stateNames = stateNames, formula = formula)
```

## 3 states with gaps

```r
# prep data for HMM
rawData <- data_test_new[,c("trip","lon","lat", "datetime", "dive", "gaps")]
colnames(rawData) <- c("ID", "lon", "lat", "datetime", "dive", "gaps")
birdData <- prepData(data=rawData, type = "LL", coordNames = c("lon", "lat"))
birdData
```

```r
# parameters initialization
## cluster K-Means for initialization
### STEP
clusterBird_step <- kmeans(na.omit(data.frame(birdData$step)), 3)
muS_1 <- max(clusterBird_step$centers)
muS_2 <- median(clusterBird_step$centers) 
muS_3 <- min(clusterBird_step$centers) 
sdS_1 <- sd(na.omit(birdData$step)[clusterBird_step[[1]] == which(clusterBird_step$centers == max(clusterBird_step$centers))])
sdS_2 <- sd(na.omit(birdData$step)[clusterBird_step[[1]] == which(clusterBird_step$centers == median(clusterBird_step$centers))])
sdS_3 <- sd(na.omit(birdData$step)[clusterBird_step[[1]] == which(clusterBird_step$centers == min(clusterBird_step$centers))])

### ANGLE
## for von mises
kappaA_1 <- 4
kappaA_2 <- 2
kappaA_3 <- 0.1

### GAPS
## beta
p1 <- 0.1
p2 <- 0.1
p3 <- 0.9

### ZERO MASS
zeroMass <- length(which(birdData$step == 0))/nrow(birdData) #we need to include zeroMass parameters
### ZERO MASS
zeroM <- length(which(birdData$gaps == 0))/nrow(birdData) #we need to include zeroMass parameters
oneM <- length(which(birdData$gaps == 1))/nrow(birdData) #we need to include zeroMass parameters

### FIT MODEL
stateNames <- c("fly","observe", "dive")
dist = list(step = "gamma", angle = "vm", gaps = "bern")
anglePar0 <- c(kappaA_1, kappaA_2, kappaA_3) 
stepPar0 <- c(muS_1, muS_2, muS_3, sdS_1, sdS_2, sdS_3, zeroMass, zeroMass, zeroMass)
gapsPar0 <- c(p1, p2, p3)
# stepPar0 <- c(muS_1, muS_2, muS_3, sdS_1, sdS_2, sdS_3) 
```

```r
m <- fitHMM(data = birdData, nbStates = 3, dist = dist,
            Par0 = list(step = stepPar0, angle = anglePar0, gaps = gapsPar0), stateNames = stateNames)
```

```r
data <- m$data
prob = stateProbs(m)
data$prediction = prob[, 'dive']
```

```r
write.table(data, file = paste0('./results/LB_hmm_', resolution, 's_2.csv'), sep = ',', row.names = FALSE)
```

## Output Resolution

```r
data_output <- data.frame()
resolution_output <- 5

for (id in unique(data$ID)){
    t <- data[data$ID == id,]
    
    idx = seq(1, nrow(t), by = resolution_output)
    traj <- t[idx,c("ID", "datetime", "x", "y")]
    
    # dives as maximum in window 
    traj$dive = sapply(idx, function(i){
            max(t$dive[i:(i+resolution_output-1)], na.rm = TRUE)
        })
    traj$prediction = sapply(idx, function(i){
            max(t$prediction[i:(i+resolution_output-1)], na.rm = TRUE)
        })
    
    
    data_output <- rbind(data_output, traj)
}
```
